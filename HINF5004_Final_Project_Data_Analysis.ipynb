```python
from scipy.stats import pearsonr, ttest_ind

import pandas as pd
```

# Method 1: DATA INTEGRAION

## STEP 1: Load the datasets

```python
# 1. USDA Food Access (Tract Level)
df_usda = pd.read_csv("Food Access Research Atlas.csv", low_memory=False)
df_usda
```

```python
# 2. CDC PLACES (County Level)
df_cdc = pd.read_csv("PLACES__Local_Data_for_Better_Health,_County_Data_2024_release_20251125.csv", low_memory=False)
df_cdc = df_cdc[df_cdc['StateDesc'] != 'United States']
df_cdc
```

```python
# 3. ACS Economic (DP03)
df_acs_econ = pd.read_csv("ACSDP5Y2023.DP03-Data.csv", header=0, low_memory=False).iloc[1:] # skip row 1 (description text)
df_acs_econ
```

```python
# 4. ACS Social (DP02)
df_acs_soc = pd.read_csv("ACSDP5Y2023.DP02-Data.csv", header=0, low_memory=False).iloc[1:] # skip row 1 (description text)
df_acs_soc
```

## STEP 2: Create a unified "FIPS" column

```python
# 1. Clean USDA FIPS
# 'CensusTract' is an integer like 1001020100, but we need 11 digits (01001020100).
# Then we take the first 5 digits (01001) for the County.
df_usda['CensusTract'] = df_usda['CensusTract'].astype(str).str.zfill(11)
df_usda['FIPS'] = df_usda['CensusTract'].str[:5]
df_usda['FIPS']
```

```python
# 2. Clean CDC FIPS
# 'LocationID' is an integer like 1001, but we need 5 digits (01001).
df_cdc['FIPS'] = df_cdc['LocationID'].astype(str).str.zfill(5)
df_cdc['FIPS']
```

```python
# 3. Clean ACS FIPS
# 'GEO_ID' is "0500000US01001," but we need the last 5 digits.
df_acs_econ['FIPS'] = df_acs_econ['GEO_ID'].str.split('US').str[-1]
df_acs_econ['FIPS']
```

```python
df_acs_soc['FIPS'] = df_acs_soc['GEO_ID'].str.split('US').str[-1]
df_acs_soc['FIPS']
```

## STEP 3: Aggregate USDA to county level

```python
# Group by FIPS and extract variables of interest.
# - 'FIPS': Federal Informaton Processing Standards
# - 'lapop1': Population count beyond 1 mile from supermarket
# - 'Pop2010': Population count from 2010 census
# - 'LILATracts_1And10': Flag for low-income and low access when considering low accessibilty at 1 and 10 miles
# - 'Urban': Flag for urban tract
county_food = df_usda.groupby('FIPS').agg({
    'lapop1': 'sum', 
    'Pop2010': 'sum',
    'LILATracts_1And10': 'max',
    'Urban': 'max'  
}).reset_index()

# 'Pct_Low_Access_Pop': Percentage of Low Access Population
county_food['Pct_Low_Access_Pop'] = (county_food['lapop1'] / county_food['Pop2010']) * 100
county_food
```

## STEP 4: Prepare CDC Health outcomes

```python
# Filter for 3 diseases: diabetes, hypertension, obesity
# - 'Age-adjusted prevalence': to compare the "healthiness" of two counties fairly, removing the bias caused by one county having an older or younger population than the other.
diseases = ['DIABETES', 'BPHIGH', 'OBESITY']
df_cdc_filtered = df_cdc[df_cdc['MeasureId'].isin(diseases) & (df_cdc['Data_Value_Type'] == 'Age-adjusted prevalence')]
df_cdc_filtered
```

```python
# Group by FIPS, State, County and extract variables of interest
# - 'FIPS': Federal Informaton Processing Standards
# - 'StateDesc': State name
# - 'LocationName': County name
# - 'MeasureId': Measure identifier (Disease)
# - 'Data_Value': Data Value, such as 14.7 (Prevalence)
county_health = df_cdc_filtered.pivot(
    index=['FIPS', 'StateDesc', 'LocationName'], 
    columns='MeasureId', 
    values='Data_Value'
).reset_index()

# Rename the columns
county_health.columns.name = None
county_health = county_health.rename(columns={
    'StateDesc': 'State',
    'LocationName': 'County',
    'DIABETES': 'Diabetes_Prev', 
    'BPHIGH': 'Hypertension_Prev', 
    'OBESITY': 'Obesity_Prev'
})

county_health
```

## STEP 5: Prepare ACS (Socioeconomic)

```python
# Extract variables of interest
# - 'FIPS': Federal Informaton Processing Standards
# - 'DP03_0062E': Median Household Income (Median)
# - 'DP03_0128PE': Percentage of families and people whose income in the past in the past 12 months is below the poverty level
acs_econ_clean = df_acs_econ[['FIPS', 'DP03_0062E', 'DP03_0128PE']].copy()

# Rename the columns
acs_econ_clean = acs_econ_clean.rename(columns={
    'DP03_0062E': 'Median_Income',
    'DP03_0128PE': 'Poverty_Rate'
})

# Convert to numeric
# 'errors=coerce' turns non-numbers (like "-") into NaN
acs_econ_clean['Median_Income'] = pd.to_numeric(acs_econ_clean['Median_Income'], errors='coerce')
acs_econ_clean['Poverty_Rate'] = pd.to_numeric(acs_econ_clean['Poverty_Rate'], errors='coerce')

acs_econ_clean
```

```python
# Extract variables of interest
# - 'FIPS': Federal Informaton Processing Standards
# - 'DP02_0068PE': Educational attainment - population 25 years and over, bachelor's degree or higher
acs_soc_clean = df_acs_soc[['FIPS', 'DP02_0068PE']].copy()
acs_soc_clean = acs_soc_clean.rename(columns={
    'DP02_0068PE': 'Bachelors_Degree_Pct'
})

# Convert to numeric
# 'errors=coerce' turns non-numbers (like "-") into NaN
acs_soc_clean['Bachelors_Degree_Pct'] = pd.to_numeric(acs_soc_clean['Bachelors_Degree_Pct'], errors='coerce')
acs_soc_clean
```

## STEP 6: Merge datasets

```python
# 1. Outer Merge: Socio + Economic Status
county_ses = pd.merge(acs_econ_clean, acs_soc_clean, on='FIPS', how='outer')
county_ses
```

```python
# 2. Inner Merge: Food + Health
master = pd.merge(county_food, county_health, on='FIPS', how='inner')
master
```

```python
# 3. Inner Merge: (Food + Health) + Socioeconomic Status
master = pd.merge(master, county_ses, on='FIPS', how='inner')
master
```

## STEP 7: Clean the data

```python
# 1. Check how many rows we have
print(f"Counties before dropping nulls: {len(master)}")
```

```python
# 2. Drop rows if they are missing critical data
master_clean = master.dropna(subset=['Pct_Low_Access_Pop', 'Diabetes_Prev', 'Hypertension_Prev', 'Obesity_Prev', 'Median_Income', 'Poverty_Rate', 'Bachelors_Degree_Pct'])

print(f"Counties after dropping nulls: {len(master_clean)}")
master_clean
```

```python
# 3. Reorder columns
cols = ['FIPS', 'State', 'County', 'Diabetes_Prev', 'Hypertension_Prev', 'Obesity_Prev', 'Pct_Low_Access_Pop', 'Median_Income', 'Poverty_Rate', 'Bachelors_Degree_Pct', 'Urban']
remaining = [c for c in master_clean.columns if c not in cols]

master_clean = master_clean[cols + remaining]
master_clean
```

# Method 2: DESCRIPTIVE ANALYSIS

```python
# 1. SAMPLE SIZE (N)
print(f"\n1. Sample Size:")
print(f"Total Counties in Analysis (N): {len(master_clean)}")
```

```python
# 2. Independent Variable (Food Access)
print(f"2. Food Insecurity (Low Access %)")
print(f"Mean: {master['Pct_Low_Access_Pop'].mean():.2f}%")
print(f"SD:   {master['Pct_Low_Access_Pop'].std():.2f}%")
```

```python
# 3. Dependent Variables (Diabetes, Hypertension, Obesity)
print(f"3. Health Outcomes")

diseases = ['Diabetes_Prev', 'Hypertension_Prev', 'Obesity_Prev']
for col in diseases:
    mean_val = master_clean[col].mean()
    sd_val = master_clean[col].std()
    print(f"{col}: Mean = {mean_val:.2f}% (SD = {sd_val:.2f})")
```

```python
# 4. Moderators (SES)
print(f"4. Socioeconomic Context")
# Income
inc_mean = master_clean['Median_Income'].mean()
inc_sd = master_clean['Median_Income'].std()
print(f"   Median Household Income: ${inc_mean:,.0f} (SD: ${inc_sd:,.0f})")

# Poverty
pov_mean = master_clean['Poverty_Rate'].mean()
pov_sd = master_clean['Poverty_Rate'].std()
print(f"   Poverty Rate: {pov_mean:.2f}% (SD: {pov_sd:.2f}%)")

# Education (The missing piece!)
edu_mean = master_clean['Bachelors_Degree_Pct'].mean()
edu_sd = master_clean['Bachelors_Degree_Pct'].std()
print(f"   Bachelor's Degree %: {edu_mean:.2f}% (SD: {edu_sd:.2f}%)")
```

# Method 3: BIVARIATE ANALYSIS

```python
# Hypothesis 1
print("1. Hypothesis 1\n")
for d in diseases:
    # Note: If Hypertension still has NaNs even in 'clean', dropna ensures safety
    clean_subset = master_clean.dropna(subset=[d]) 
    
    r, p = pearsonr(clean_subset['Pct_Low_Access_Pop'], clean_subset[d])
    
    # Format p-value string
    p_text = "< 0.001" if p < 0.001 else f"= {p:.3f}"
    sig_text = "**SIGNIFICANT**" if p < 0.05 else "Not Significant"
    
    print(f"{d} vs. Food Access (N={len(clean_subset)}):")
    print(f"   r = {r:.3f}")
    print(f"   p {p_text}  --> {sig_text}")
    print("-" * 30)
```

# Method 4: STRATIFIED ANALYSIS

```python
# Hypothesis 2: Income
print("2-1. Hypothesis 2: Income")

for d in diseases:
    # Split income into 'High Income Group' and 'Low Income Group'
    median_inc = master_clean['Median_Income'].median()
    low_inc = master_clean[master_clean['Median_Income'] < median_inc]
    high_inc = master_clean[master_clean['Median_Income'] >= median_inc]

    print(f"\nAverage {d}:")
    print(f"Low Income Group:  {low_inc[d].mean():.2f}%")
    print(f"High Income Group: {high_inc[d].mean():.2f}%")

    # Test: Does the Food-Disease link disappear in low income counties?
    r_low, _ = pearsonr(low_inc['Pct_Low_Access_Pop'], low_inc[d])
    r_high, _ = pearsonr(high_inc['Pct_Low_Access_Pop'], high_inc[d])

    print(f"\nCorrelation:")
    print(f"Low Income Group:   r = {r_low:.3f}")
    print(f"High Income Group:  r = {r_high:.3f}")
    print("-" * 30)
```

```python
# Hypothesis 2: Poverty
print("2-2. Hypothesis 2: Poverty")

for d in diseases:
    # Split income into 'High Poverty Group' and 'Low Poverty Group'
    median_pov = master_clean['Poverty_Rate'].median()
    low_pov = master_clean[master_clean['Poverty_Rate'] < median_pov]
    high_pov = master_clean[master_clean['Poverty_Rate'] >= median_pov]

    print(f"\nAverage {d}:")
    print(f"Low Poverty Group:  {low_pov[d].mean():.2f}%")
    print(f"High Poverty Group: {high_pov[d].mean():.2f}%")

    # Test: Does the Food-Disease link disappear in low poverty counties?
    r_low, _ = pearsonr(low_pov['Poverty_Rate'], low_pov[d])
    r_high, _ = pearsonr(high_pov['Poverty_Rate'], high_pov[d])

    print(f"\nCorrelation:")
    print(f"Low Poverty Group:   r = {r_low:.3f}")
    print(f"High Poverty Group:  r = {r_high:.3f}")
    print("-" * 30)
```

```python
# Hypothesis 2: Education
print("2-3. Hypothesis 2: Education")

for d in diseases:
    # Split income into 'High Education Group' and 'Low Education Group'
    median_edu = master_clean['Bachelors_Degree_Pct'].median()
    low_edu = master_clean[master_clean['Bachelors_Degree_Pct'] < median_edu]
    high_edu = master_clean[master_clean['Bachelors_Degree_Pct'] >= median_edu]

    print(f"\nAverage {d}:")
    print(f"Low Education Group:  {low_edu[d].mean():.2f}%")
    print(f"High Education Group: {high_edu[d].mean():.2f}%")

    # Test: Does the Food-Disease link disappear in low education counties?
    r_low, _ = pearsonr(low_edu['Bachelors_Degree_Pct'], low_edu[d])
    r_high, _ = pearsonr(high_edu['Bachelors_Degree_Pct'], high_edu[d])

    print(f"\nCorrelation:")
    print(f"Low Education Group:   r = {r_low:.3f}")
    print(f"High Education Group:  r = {r_high:.3f}")
    print("-" * 30)
```

# Method 5: GROUPED COMPARISION

```python
# Hypothesis 3
print("Hypothesis 3: Rural vs Urban\n")

# 0 = Rural, 1 = Urban
rural = master_clean[master_clean['Urban'] == 0]
urban = master_clean[master_clean['Urban'] == 1]

# for each disease, calculate average prevalence in rural/urban, t-test p-value and derive result
for d in diseases:
    print(f"\n--- {d} ---")
    mean_rural = rural[d].mean()
    mean_urban = urban[d].mean()
    
    print(f"   Avg Prevalence (Rural): {mean_rural:.2f}% (N={len(rural)})")
    print(f"   Avg Prevalence (Urban): {mean_urban:.2f}% (N={len(urban)})")
    
    # Independent T-Test
    t_stat, p_val = ttest_ind(rural[d], urban[d], nan_policy='omit', equal_var=False)
    
    print(f"   T-Test p-value: {p_val:.3f}")
    if p_val < 0.05:
        print("   Result: SIGNIFICANT Difference")
    else:
        print("   Result: No Significant Difference")
```
